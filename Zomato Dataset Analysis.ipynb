### Zomato Data Analysis and Visualization

### Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('dark_background')
```

### Reading CSV

```python
df = pd.read_csv('zomato.csv')
df.head()
```

```python
df.shape  # 17 columns and 52K rows
```

```python
df.columns
```

### Dropping Columns

```python
# df = df.drop(columns = ['url', 'address', 'phone', 'menu_item', 'dish_liked', 'reviews_list'])

df = df.drop(['url', 'address', 'phone', 'menu_item', 'dish_liked', 'reviews_list'], axis = 1)
df.head()
```

```python
df.shape   # Only 11 columns now
```

```python
df.info()
```

* There are some Null values in **'rate'** column.

### Dropping Duplicates

```python
df.drop_duplicates(inplace = True)

df.shape
```

### Cleaning 'rate' Column

```python
df['rate'].unique()
```

#### Removing "NEW", "-" and "/5" from 'rate' Column

```python
def handlerate(value):
    if(value=='NEW' or value=='-'):
        return np.nan                      # replace 'NEW' and '-' with Null value using np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
    
df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()
```

### Filling Null Values in 'rate' Column with Mean

```python
# To find out the number of null values in 'rate' column

df['rate'].isnull().sum()  
```

```python
# Filling null values with mean of the rate column

df['rate'].fillna(df['rate'].mean(), inplace = True)

df.rate.isnull().sum()
```

```python
df.info()
```

> *Null Values in columns **location**, **cuisines**, **rest_type**. But they are not many, so we can drop them.*

### Dropping Null Values

```python
df.dropna(inplace = True)

df.head()
```

### Renaming Columns

```python
df.rename(columns = {'approx_cost(for two people)':'Cost2plates', 'listed_in(type)':'Type'}, inplace = True)

df.head()
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

### listed_in(city) and location, both are there, lets keep only one.

```python
# dropping listed_in(city) column

df = df.drop(['listed_in(city)'], axis = 1)
```

```python
df.head()
```

### Cleaning 'Cost2plates' Column

```python
df['Cost2plates'].unique()
```

> *Values are of string type with commas.*

#### Removing ',' from 'Cost2Plates' Column

```python
def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',', '')
        return float(value)
    else:
        return float(value)
    
df['Cost2plates'] = df['Cost2plates'].apply(handlecomma)
df['Cost2plates'].unique()
```

```python
df.head()
```

### Cleaning 'rest_type' Column

```python
df['rest_type'].unique()
```

```python
df['rest_type'].value_counts()
```

> *Clearly, Some of the restraunt types are very few. Let's cluster them into one group.*

```python
rest_types = df['rest_type'].value_counts(ascending = False)

rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types < 1000]

rest_types_lessthan1000
```

#### Making rest_types less than 1000 in frequency as others

```python
def handle_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value
    
df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
df.head()
```

### Cleaning 'location' Column

```python
location = df['location'].value_counts(ascending  = False)
location
```

```python
location_lessthan300 = location[location<300]

location_lessthan300
```

```python
# Clustering area with less than 300 restaurants as other location

def handle_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value
        
df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
df.head()
```

### Cleaning 'cuisines' Column

```python
cuisines = df['cuisines'].value_counts(ascending  = False)
cuisines
```

> *There are few Restaurants for some combination of Cuisine.*

```python
# Clustering Restraunt with unique cuisine into other

cuisines_lessthan100 = cuisines[cuisines<100]

def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value
        
df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```

```python
df.head()
```

### Checking 'Type' column

```python
df['Type'].value_counts()
```

> *Good amount of numbers, so can't cluster them.*

### Visualizing Top Cuisines

> *Top cuisines for restaurants would be North Indian, North Indian + chinese or South Indian in bangalore*
